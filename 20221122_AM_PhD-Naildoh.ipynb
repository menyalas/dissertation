```r
# Functions
factorize <- function(df){ # Create a function
  for(i in which(sapply(df, class) == "character")) # that looks for variables with the character class 
      df[[i]] = as.factor(df[[i]]) # and converts them to factor (i.e., categorical) class
  return(df)
}

unfactorize <- function(df){ # Create a function
  for(i in which(sapply(df, class) == "factor")) # that looks for variables with the character class 
      df[[i]] = as.character(df[[i]]) # and converts them to factor (i.e., categorical) class
  return(df)
}
```

On first glance, I see problems, for example with some of these being memoirs rather than complete autobiographies (e.g., "Twelve Years of My Life"). Also, I expect some of these are exerpts or something other than a complete autobiography because multiple authors are represented in single sources (e.g., America's Immigrants: Adventures in Eyewitness Hisory). I am leaning toward using -- as a starting point -- a vetted bibliography, such as the list of working class autobiographies included in <i>Bread, knowledge, and freedom: A study of nineteenth-century working class autobiography</i> by David Vincent (1981). These conservatively cover the time period in question. Another possibility is the <i>The autobiography of the working class: An annotated, critical bibliography</i> by John Burnett, David Vincent and David Mayall (1984), but this collection extends well into the 21st century. In both cases, these are British autobiographies, so they do not cover the geographical area of my study, which is the United Kingdom and its uncontested settler colonies (U.S., Canada, Australia, New Zealand) per Burke 2000) during the long 19th century (1750-1914). For now, I will just add letters and diaries to the corpus.
