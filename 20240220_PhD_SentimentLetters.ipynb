```python
from nltk.corpus import PlaintextCorpusReader
```

```python
import pandas as pd
```

```python
import re
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sentimentAnalyzer = SentimentIntensityAnalyzer() #Initialize sentiment scorer
```

```python
lettersCorpus = PlaintextCorpusReader('letters', '.*txt') #Define corpus
len(lettersCorpus.fileids()) #How many files
```

```python
sentimentLetter = pd.DataFrame(columns = ['text','docid','scoreNeg', 'scorePos', 'scoreNeu', 'scoreCompound']) # Create a new dataframe to hold sentences, letter id and compound scores
sentimentLetter.head()
```

```python
for fileid in lettersCorpus.fileids(): # For each file in the corpus
    f = open("letters/cleaned/" + fileid, "w")
    text = lettersCorpus.raw(fileid) # Place the string into the object "text"
    text = re.sub(r"(\. ){2,}","", text) # Remove ellipses
    text = re.sub(r"(\n+)"," ", text) # Replace blank lines with a single space
    text = re.sub(r"(&dot)","", text) # Remove this expression (a dot)
    text = re.sub(r"(— )","", text) # Remove dashes
    text = re.sub(r"(\r)"," ", text) # Replace /r with blank space
    text = re.sub(r"(\˙)","", text) # Replace dot with no space
    scores = sentimentAnalyzer.polarity_scores(text) # Calculate sentiment scores
    new_row = pd.Series([text, fileid, scores['neg'], scores['pos'], scores['neu'], scores['compound']], index=["text", "docid", "scoreNeg", "scorePos", "scoreNeu", "scoreCompound"])
    sentimentLetter = pd.concat([sentimentLetter, new_row.to_frame().T], ignore_index=True)
    f.write(text)
    f.close()
    print(fileid) # Show progress
```

```python
sentimentLetter
```

```python
# Remove the .txt from the file name
sentimentLetter['docid'] = sentimentLetter['docid'].str.replace(r'.txt', '', regex=True)
sentimentLetter
```

```python
sentimentLetter.to_csv('20240220_PhD_SentimentLetter.csv', index = True)
```
